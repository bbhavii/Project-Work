```python
# Program to generate a random number between 0 and 9

# importing the random module
import random
import matplotlib.pyplot as plt
import pandas as pd
x = [0]*1000
y = [0]*1000

#print(x)

for i in range(0,250):
  x[i]=random.randint(0,30)
  y[i]=random.randint(0,47)

for i in range(250,500):
  x[i]=random.randint(50,80)
  y[i]=random.randint(70,100) 

for i in range(500,800):
  x[i]=random.randint(10,40)
  y[i]=random.randint(54,90) 

for i in range(800,1000):
  x[i]=random.randint(0,100)
  y[i]=random.randint(0,100) 

plt.scatter(x,y);

Data = pd.DataFrame(x,columns=['x'])
Data['y'] = y
display(Data)

Data.to_csv('Data.csv')

from IPython.display import HTML
import pandas as pd
import numpy as np



def create_download_link(title = "Download CSV file", filename = "data.csv"):  
    html = '<a href={filename}>{title}</a>'
    html = html.format(title=title,filename=filename)
    return HTML(html)

# create a link to download the dataframe which was saved with .to_csv method
create_download_link(filename='Data.csv')
```
